```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
feature = pd.read_csv('features.csv')
stores = pd.read_csv('stores.csv')
data = pd.read_csv('train.csv')
```

```python
store_data = data.merge(stores , on = 'Store' )
store_data.drop('IsHoliday' , axis=1 , inplace = True)

sales = store_data.merge(feature , on = ['Store' , 'Date'])
```

```python
sales.head()
```

```python
sales['IsHoliday'] = np.where(sales['IsHoliday']==True , 1,0)

sales['MarkDown1']= np.where(sales['MarkDown1'].isnull() , 0 ,sales['MarkDown1'])
sales['MarkDown2']= np.where(sales['MarkDown2'].isnull() , 0 ,sales['MarkDown2'])
sales['MarkDown3']= np.where(sales['MarkDown3'].isnull() , 0 ,sales['MarkDown3'])
sales['MarkDown4']= np.where(sales['MarkDown4'].isnull() , 0 ,sales['MarkDown4'])
sales['MarkDown5']= np.where(sales['MarkDown5'].isnull() , 0 ,sales['MarkDown5'])

temp = pd.get_dummies(sales['Type'] )
sales = pd.concat([sales , temp] , axis = 1)

sales.drop(['Date' , 'Type'] , axis = 1 , inplace = True)
```

```python
sales.tail(20)
```

```python
cor = sales.corr()
```

```python
from sklearn.model_selection import train_test_split

feature_train , feature_test ,label_train , label_test = train_test_split(sales.drop(columns='Weekly_Sales') , 
                                                                           sales['Weekly_Sales'] ,
                                                                           test_size = 0.3,
                                                                           random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
#model= LinearRegression()
model =RandomForestRegressor()
model.fit(feature_train , label_train)
predict = model.predict(feature_train)

from sklearn.metrics import r2_score

print('train : ' , r2_score(label_train , predict))
predict_test = model.predict(feature_test)
print('test : ' , r2_score(label_test , predict_test))
from sklearn import metrics

print('rmse : trian = ' , np.sqrt(metrics.mean_squared_error(label_train , predict)))
print('rmse: test = ' , np.sqrt(metrics.mean_squared_error(label_test , predict_test)))
```
